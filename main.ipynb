```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_percolation(L, p):
    
      matrix = np.random.choice([0, 1], size=(L, L), p=[1 - p, p])
      visited = set()

      def dfs(x, y):
          stack = [(x, y)]

          while stack:
              x, y = stack.pop()
              if x < 0 or x >= L or y < 0 or y >= L:
                  continue
              if matrix[x, y] == 0 or (x, y) in visited:
                  continue
              if y == L - 1:
                  return True

              visited.add((x, y))

              
              stack.extend([(x+1, y), (x-1, y), (x, y+1), (x, y-1), (x-1, y-1), (x-1, y+1), (x+1, y-1), (x+1, y+1)]) #neighbors

      for i in range(L):
          if matrix[i, 0] == 1 and (i, 0) not in visited and dfs(i, 0):
              return True

      return False

L_values = [20, 50, 100]
p_values = np.linspace(0, 1, 20)  #Number of different p to check

# Run simulations for different L and p values
percolation_probabilities_all = []

for L in L_values:
    percolation_probabilities_L = []
    for p in p_values:
        percolation_count = 0
        total_simulations = 1000  #Number of simulations

        for _ in range(total_simulations):
            if simulate_percolation(L, p):    #if fire hits add 1
                percolation_count += 1

        percolation_probabilities_L.append(percolation_count / total_simulations)

    percolation_probabilities_all.append(percolation_probabilities_L)

# Plot the results
for i, L in enumerate(L_values):
    plt.plot(p_values, percolation_probabilities_all[i], label=f'L = {L}')

plt.xlabel('Probability p of placing a tree')
plt.ylabel('Probability that firehits the opposite edge of the grid during the simulation')
plt.legend()
plt.show()

```

```python

def hoshen_kopelman(matrix):
    labels = np.zeros_like(matrix, dtype=int)
    next_label = 1

    def get_label(x, y): #root label of a cluster using  Hoshen-Kopelman
        while labels[x, y] < 0:
            x, y = -labels[x, y] // L, -labels[x, y] % L
        return labels[x, y]

    for i in range(L):
        for j in range(L):
            if matrix[i, j] == 1:
                neighbors = []
                if i > 0 and matrix[i - 1, j] == 1:
                    neighbors.append(get_label(i - 1, j))
                if j > 0 and matrix[i, j - 1] == 1:
                    neighbors.append(get_label(i, j - 1))

                if not neighbors:
                    labels[i, j] = next_label
                    next_label += 1
                else:
                    min_label = min(neighbors)
                    labels[i, j] = min_label
                    for neighbor in neighbors:
                        if neighbor != min_label:
                            labels[labels == neighbor] = min_label

    cluster_sizes = np.bincount(labels.ravel())
    cluster_sizes[0] = 0  

    return cluster_sizes

def simulate_percolation(L, p):
    matrix = np.random.choice([0, 1], size=(L, L), p=[1 - p, p])
    cluster_sizes = hoshen_kopelman(matrix)

    return np.max(cluster_sizes)

L = 100
num_simulations = 100
p_values = np.linspace(0, 1, 100)  

# Simulate percolation for different p values
average_sizes = []

for p in p_values:
    sizes = [simulate_percolation(L, p) for _ in range(num_simulations)]
    average_sizes.append(np.mean(sizes))

# Plot the results
plt.plot(p_values, average_sizes)
plt.title('Average Size of Largest Cluster vs. p')
plt.xlabel('probability p of placing a tree')
plt.ylabel('Average Size of Largest Cluster')
plt.show()
```
